# Multi-Function Calculator

```python
import sympy
import math
from sympy import symbols, solve

def multi_calc():
    print("Multi-Function Calculator Menu: \n",
            "1. Addition                     8. Solve for X\n",
            "2. Subtraction                  9. Solve Proportions\n",
            "3. Multiplication              10. Factor Square Roots\n",
            "4. Division                    11. Convert Decimals TO Fractions and Percents\n",
            "5. Detect Prime Number         12. Convert Fractions TO Decimals and Percents\n",
            "6. Generate Prime Factors      13. Convert Percents TO Decimals and Fractions\n",
            "7. Simplify Square Roots\n")


    while True:
        user_choice = int(input("Select a number(1-13) from the menu: "))
        
        if user_choice==1:
            def add():
                a = int(input("What number do you want to add?: "))
                b = int(input("Enter the second number to be added: "))
                print("The sum is:", a + b)
            add()
            break

        elif user_choice==2:
            def sub():
                c = int(input("What number do you wish to be to be subtracted?: "))
                d = int(input("Enter the second number to be subtracted from: "))
                print("The difference is:", c - d)
            sub()
            break

        elif user_choice==3:
            def mult():
                e = int(input("What number do you want to multiply?: "))
                f = int(input("Enter the second number to be multiplied: "))
                print("The product is:", e * f)
            mult()
            break

        elif user_choice==4:
            def div():
                g = int(input("What number do you want to be divided(the dividend)?: "))
                h = int(input("Enter the second number to be divided by(the divisor): "))
                print("The quotient is:", g / h)
            div()
            break

        elif user_choice==5:
            def prime_n():
                number = int(input("Is this number prime or not?: "))

                prime_or_comp = "a prime number"

                for test_number in range(2,number):
                    if number % test_number==0:
                        prime_or_comp = "a composite number"

                    print(number, "is", prime_or_comp)
            prime_n()
            break

        elif user_choice==6:
            def prime_facs():
                num = int(input("What number do you want to find all the prime factors for?: "))
                prime_nums = sympy.primefactors(num)
                print("The prime factors of", num, "are/is:", prime_nums)
            prime_facs()
            break

        elif user_choice==7:
            def sq_rt(): 
                n = int(input('What number do you wish to find the square root of?: '))

                upper_limit = math.floor(math.sqrt(n)) + 1
                max_factor = 1
                other_factor = 1
                square_root = 1

                # loop to find greatest sq factor
                for maybe_factor in range(1, upper_limit):
                    if n % (maybe_factor**2) == 0:
                        max_factor = maybe_factor**2

                # divide out the greatest sq factor
                other_factor = n/max_factor

                #output
                square_root = int(math.sqrt(max_factor))
                other_factor = int(other_factor)
                print("The square root of", n, "is:", square_root*sympy.sqrt(other_factor))
            sq_rt()
            break

        elif user_choice==8:
            def solv():
                x = symbols('x')

                #input equation
                eq = input('What equation do you want to solve for x?: 0 = ')

                #solve equation
                solution = solve(eq,x)
                print("x =", solution[0])
            solv()
            break

        elif user_choice==9:
            def solve_props():
                n1 = int(input("Enter numerator of the first proportion. Enter zero if unknown value: "))
                d1 = int(input("Enter denominator of the first proportion. Enter zero if unknown value: "))
                n2 = int(input("Enter numerator of the second proportion. Enter zero if unknown value: "))
                d2 = int(input("Enter denominator of the second proportion. Enter zero if unknown value: "))

                if n1==0:
                    answer = n2 * d1 / d2
                    print("n1 = ", answer)

                elif d1==0:
                    answer = n1 * d2 / n2
                    print("d1 = ", answer)

                elif n2==0:
                    answer = d2 * n1 / d1
                    print("n2 = ", answer)

                elif d2==0:
                    answer = n2 * d1 / n1
                    print("d2 = ", answer)
            solve_props()
            break

        elif user_choice==10:
            def factor_sq():
                n = int(input('Enter a square root to factor: '))

                upper_limit = math.floor(math.sqrt(n)) + 1
                max_fact = 1
                other_fact = 1
                sq_rt = 1

                for mf in range(1, upper_limit):
                    if n % (mf **2) == 0:
                        max_fact = mf**2

                other_fact = n/max_fact

                sq_rt = int(math.sqrt(max_fact))
                other_fact = int(other_fact)
                factors = sq_rt*sympy.sqrt(other_fact)

                #output
                print(factors)
            factor_sq()
            break

        elif user_choice==11:
            def dec_2_frac_perc():
                decimal = input("Enter a decimal number to convert: ")

                # get number of decimal places as an integer
                exponent = int(len(decimal))-1

                # Convert the input to a float number
                n = float(decimal)

                #use exponent to get the numerator
                num = int(n * 10**exponent)

                # useexpoent to get the denominator
                den = 10**exponent

                # convert to percents
                percent = n * 100

                #output of fracton and percents
                print(f"{decimal} converts to: {num}/{den} and {percent:,.2f}%")
            dec_2_frac_perc()
            break

        elif user_choice==12:
            def frac_2_dec_perc():
                # enter fraction
                n = int(input("Enter the numerator: "))
                d = int(input("Enter the denominator: "))

                #convert to decimal and percent
                decimal = n/d
                percent = (n/d)* 100

                #output
                print(f"{n}/{d} converts to: {decimal:,.4f} and {percent:,.2f}%")
            frac_2_dec_perc()
            break

        elif user_choice==13:
            def perc_2_dec_frac():
                #percent input
                percent = input("Enter percent to be converted: ")

                #convert to decimal
                dec = float(percent)/100

                #convert to fraction
                exponent = int(len(percent))-1

                num = int(dec * 10**exponent)

                den = 10**exponent

                print(f"{percent}% converts to: {dec:,.4f} and {num}/{den}")

            perc_2_dec_frac()
            break

        else: 
            print("Incorrect entry. MUST enter a number from 1 to 13")

multi_calc()
```
